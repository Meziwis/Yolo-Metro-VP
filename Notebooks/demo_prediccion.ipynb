```python
from ultralytics import YOLO 
```

### Modelos 

Hay varios modelos que se pueden usar, acá dejo uno de detección y otro de segmentación

```python
detector_model = "yolov8n.pt"
segmentation_model = "yolov8n-seg.pt"

model = YOLO(detector_model)
```

Así podemos imprimir el nombre de las clases

```python
print(model.names)
```

### Ahora definimos algunas variables importantes para la predicción

```python
path = "../Videos/cam_6_hospital_2/cam_6_hosp_2.avi"
guardar_resultados = True
clase_interes = 6
confianza_minima = 0.25
guardar_salida = True
guardar_confianza = True
```

```python
results_video = model.predict(source=path,save=guardar_resultados,
         classes= clase_interes, conf= confianza_minima, save_txt = guardar_salida,
          save_conf = guardar_confianza)
```
